```python
import datetime as dt
import time

def arrange_dates(start_date, end_date):
    """Used in the get_election_tweets func. """
    dates = []
    delta = end_date - start_date
    
    for i in range(delta.days + 1):
        day = start_date + timedelta(days=i)
        dates.append(day)
    
    return dates

def get_election_tweets(tweets_per_day=5000, start_date="2020-10-01", end_date="2020-11-03"):
    
    # get a list of dates
    f = "%Y-%m-%d"
    sd = dt.datetime.strptime(start_date, f)
    ed = dt.datetime.strptime(start_date, f)
    print(sd, ed)
    dates = arrange_dates(sd, ed)
    
    candidates = ['Biden', 'Trump']
    
    # fetch tweets & save CSVs (each CSV contains one day/candidate)
    
    lam_plus_1 = lambda x: (x + timedelta(days=1)).strftime("%Y-%m-%d")
    i = 0 # will use for sleep timer to avoid 'client payload' error'
    
    for date in dates:
        for candidate in candidates:
            
            d = date.strftime("%Y-%m-%d")            
            print(f"\n Begin scraping {candidate}'s tweets on {d}")
            outfile = f"./code/pickle/tweets/{candidate}_{d}.csv"
            
            # twint search query details
            c = twint.Config()
            c.Search = candidate
            c.Lang = 'en'
            c.Since = d
            c.Until = lam_plus_1(date)
            c.Limit = tweets_per_day
            c.Store_csv = True
            c.Output = outfile
            
            # run the search
            twint.run.Search(c)
            print(f"{outfile} created.")
            time.sleep(30 + i/2) # to avoid 'client payload' error
            i += 0.5
```
