```python
from datetime import date, timedelta

def arrange_dates(start_date, end_date):
    """Used in the get_election_tweets func. """
    start_date = date(2020,10,1)
    end_date = date(2020,11,3)
    dates = []
    delta = end_date - start_date
    
    for i in range(delta.days + 1):
        day = start_date + timedelta(days=i)
        dates.append(day)
    
    return dates

def get_election_tweets(tweets_per_day=5000):
    # get a list of dates
    start_date = date(2020,10,1)
    end_date = date(2020,11,3)
    dates = arrange_dates(start_date, end_date)
    candidates = ['Biden', 'Trump']
    
    lam_plus_1 = lambda x: (x + timedelta(days=1)).strftime("%Y-%m-%d")
    
    for date in dates:
        for candidate in candidates:
            d = date.strftime("%Y-%m-%d")
            outfile = f".code/pickle/tweets/{candidate}_{d}.p"
            # twint search query details
            c = twint.Config()
            c.Search = candidate
            c.Lang = 'en'
            c.Since = d
            c.Until = lam_plus_1(date)
            c.Limit = tweets_per_day
            c.Pandas = True
            
            df = twint.storage.panda.Tweets_df
            with open(outfile, "wb") as p:
                pickle.dump(df, p)

            twint.run.Search(c)
    
```
