```python
import pandas as pd
import os
import pickle

pd.options.display.max_columns = 40
```

```python
tweets = os.listdir("../data/tweets/")
files = [csv for csv in tweets if csv[-4:] == '.csv' and csv[:3] != 'ALL']
```

```python
# combine all CSVs into one DF
dfs = []
for file in files:
    dfs.append(pd.read_csv(f"../data/tweets/{file}"))

df_t = pd.concat(dfs)

df_t.head(3)
```

```python
df_t.tweet.iloc[29843]
```

```python
df_t.drop_duplicates(subset='id', inplace=True)
```

```python
df_t.shape
```

```python
df_t.head(3)
```

```python
biden = df_t.tweet.str.contains(r"[Bb][Ii][Dd][Ee][Nn]")
df_t['biden'] = biden
trump = df_t.tweet.str.contains(r"[Tt][Rr][Uu][Mm][Pp]")
df_t['trump'] = trump
biden.value_counts(), trump.value_counts()
```

```python
df_t.to_pickle("pickle/df_t_raw.pick")
```
