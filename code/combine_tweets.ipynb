```python
import pandas as pd
import os



pd.options.display.max_columns = 40
```

```python
tweets = os.listdir("../data/tweets/")
files = [csv for csv in tweets if csv[-4:] == '.csv']
```

```python
# combine all CSVs into one DF
dfs = []
for file in files:
    dfs.append(pd.read_csv(f"../data/tweets/{file}"))

df_t = pd.concat(dfs)

df_t.head(3)
```
